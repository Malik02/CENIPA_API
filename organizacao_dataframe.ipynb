# Framework sobre acidentes aeronaúticos

## Bibliotecas

```python
import pandas as pd # importando a biblioteca pandas como a sigla pd
```

## Leitura dos Datasets 

```python
df_aeronave = pd.read_csv('./Data/aeronave.csv', sep=';') # leitura do arquivo csv a partir da biblioteca pandas e o comando read_csv
df_aeronave # chamada da várivael onde foi armaenado o dataframe
```

```python
df_fator_contribuinte = pd.read_csv('./Data/fator_contribuinte.csv', sep=';')
df_fator_contribuinte
```

```python
df_ocorrencia_tipo = pd.read_csv('./Data/ocorrencia_tipo.csv', sep=';')
df_ocorrencia_tipo
```

```python
df_ocorrencia = pd.read_csv('./Data/ocorrencia.csv', sep=';')
df_ocorrencia
```

```python
df_recomendacao = pd.read_csv('./Data/recomendacao.csv', sep=';')
df_recomendacao
```

## Juntando Datasets

```python
df_1 = df_aeronave.merge(df_ocorrencia, left_on='codigo_ocorrencia2', right_on='codigo_ocorrencia2')
df_1
```

```python
df_2 = df_1.merge(df_fator_contribuinte, left_on='codigo_ocorrencia3', right_on='codigo_ocorrencia3')
df_2
```

```python
df_3 = df_2.merge(df_ocorrencia_tipo, left_on='codigo_ocorrencia1', right_on='codigo_ocorrencia1')
df_3
```

```python
df_4 = df_3.merge(df_recomendacao, left_on='codigo_ocorrencia4', right_on='codigo_ocorrencia4')
df_4
```

```python
df_4.to_csv('./Data/df_final_aeronautico.csv')
```
